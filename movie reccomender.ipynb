```python
#loading the data from the csv movie file to pandas database
movies_data=pd.read_csv('/content/movies.csv')
```

```python
#printing the first 5 rows of the data set
movies_data.head()
```

```python
#number of rows and columns in the data frame
movies_data.shape
```

```python
#selecting the relavant features for recommendation
selected_features=('genres','keywords','tagline','cast','director')
print(selected_features)
```

see now we are replacing the null values with blank
see in fake news we dodnt give for loop because we changed our datat intially only to required columns
but here we are nullyifing only in some the columns but we still need other volumns the selected features are the deciding factors but other columns are also imp


```python
#replacing the missing value with null

for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combing all the 5 columns
combined_features=movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

cosine similarity

```python
#getting the similarities scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user

movie_name = input('Enter your favorite movie name: ')
```

```python
#creating a list of all the movie names given in the data set

list_of_all_titles= movies_data['title'].tolist()
print(list_of_all_titles)
```

we are comparing the dta given by the user and printing the close matches to the data from the data set

```python
#finding the cosed max for the movie name entered
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

we are printing the first closest matching value

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title

index_of_the_movie = movies_data.title == close_match
print(index_of_the_movie)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

now what we are doing is:


```python
#getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score

sorted_simialr_movies=  sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_simialr_movies)
```

```python
#print the name of similar movies based on the index

print('Movies suggested for you: \n')
```

```python
i=1
for movie in sorted_simialr_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```
